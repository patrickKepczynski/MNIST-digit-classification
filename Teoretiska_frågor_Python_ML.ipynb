### 16. Förklara (gärna med ett exempel): Ordinal encoding, one-hot encoding, dummy variable encoding.

Svar: 

*Kategorisk data är oftast i textform och alltså inte numerisk. För att kunna använda datan i en ML-modell behöver vi transformera den på olika sätt.*

*Ordinal encoding ger kategorierna rang (ifall det finns ett syfte). Det kan vara att en svart produkt (rang 0) är mer populär än en vit (rang 1) och kan på så sätt bidra till en bättre prisprediktion för produkten.*

*One-hot encoding (OHE) ger kategorierna en varsin kolumn, där med hjälp av get_dummies(), med värdet 0 eller 1 (falskt eller sant). Här skulle en blå tröja få en etta och andra färger i samma rad en nolla. Används inom alla modeller förutom linjära regressionsmodeller.*

*Variable encoding liknar OHE men vi klipper bort den första kolumnen med drop_first = True. Det räcker för datorn att förstå eftersom i de fallen där alla kategorierna ha 0 i värde, vet den att det är den bortklippta kategorin som har fått 1 i värde. Används inom linjära regressionsmodeller.*

### 19. Förklara skillnaden mellan overfitting och underfitting i en maskininlärningsmodell. Beskriv även hur man kan upptäcka respektive åtgärda dem.

Svar: 

*Modeller som är bra på träningsdatan men som presterar dåligt på valideringsdata kallas för överanpassade (overfitted). Modeller som är för enkla för att fånga strukturen i träningsdatan kallas för underanpassade (underfitted), de presterar generellt dåligt på både tränings- och valideringsdata.*

*Vi kan upptäcka overfitting ifall det finns en stor skillnad på felmarginal (RMSE) mellan träningsdata och valideringsdata. Vi kan åtgärda detta med hjälp av en enklare modell alternativt Ridge resp. Lasso regression, då komplexa modeller lätt blir överanpassade.*

*Vi kan upptäcka underfitting ifall vi har höga värden på RMSE i både träning- och valideringsdata. Vi kan åtgärda detta genom att göra modellen lite mer komplex. Man kan använda fler features (t.e.x. med polynomvariabler).*
